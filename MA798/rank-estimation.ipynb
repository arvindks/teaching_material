```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n = 512
x = np.linspace(0,1,n)
XX,YY = np.meshgrid(x,x)

f, axarray = plt.subplots(2,2,sharex=True,sharey=True)

slst = np.array([ [0.5, 0.1], [0.05, 0.01]])
for i in range(2):
    for j in range(2):
        A = np.exp(-np.abs(XX-YY)/slst[i,j])
        u,s,vh = np.linalg.svd(A)

        ell = 40
        X = np.random.randn(n, ell)/np.sqrt(ell)
        _, sX, _ = np.linalg.svd(A @ X)

        Y = np.random.randn(2*ell, n )/np.sqrt(2*ell)
        _, sY, _ = np.linalg.svd(Y @ (A @ X))


        axarray[i,j].semilogy(range(ell), s[:ell], 'k-', range(ell), sX, 'r--', range(ell), sY, 'g--')
        if i == 0 and j == 0: 
            axarray[i,j].legend(('True', 'Sketched1', 'Sketched2'))
```
