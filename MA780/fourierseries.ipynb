```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Fourier series

```python
# First plot the square wave function

x = np.linspace(0,2*np.pi,1000)
f = 0*x
f[np.flatnonzero(x <= np.pi)] = 1.

plt.plot(x,f,'k-', lw =2.)
plt.xlabel('x', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16.)
plt.title('Square wave function', fontsize = 18)
```

The Fourier series for this square wave is given by 
$$ f(x) = \frac{1}{2} + \sum_{k=1, \text{odd}}^\infty\frac{2\sin(kx)}{k\pi} \qquad x \in [0,2\pi).$$
                                                                              
We show the convergence of the partial Fourier series. 

```python
nlst = [1,5,10,100]

fig,axarray = plt.subplots(2,2,sharex=True,sharey = True, figsize= (10,8))
for n,ax in zip(nlst,axarray.flatten()):
    fn = 0*x + 1/2.
    for k in np.arange(0,n,2):
        fn += 2*np.sin((k+1)*x)/(np.pi*(k+1))
    ax.plot(x,f,'k-', x, fn,'r--', lw =2)
    ax.set_title('n = ' + str(n), fontsize = 16)
    
```
