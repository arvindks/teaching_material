```python
import matplotlib.pyplot as plt
import numpy as np
from gersh import *
```

### Localization using Gershgorin circles

Consider the matrix 

$$ A = \begin{bmatrix} 1 & 0.5 & 1 \\ 0 & 3 & -0.5 \\ 0 & 0 & 5.0 \end{bmatrix}.$$ 

This matrix has eigenvalues 1,3,-5. The centers of the discs are $1,3,-5$ and the radii are $1.5, 3.5, 0$.  

```python
A = np.array([[1.,0.5,1],[0,3.,-0.5],[0.,0,5.]])
plot_gersh(A)
```

### Some variations:


1. Not every circle needs to have its own eigenvalue (example from mathstackexchange)

$$ A = \begin{bmatrix} 1 & -1 & 0 \\ 2 & -1 & 0 \\ 0 & 1 & 4 \end{bmatrix}.  $$ 

```python
A = np.array([[1.,-1.,0.],[2.,-1.,0.],[0.,1.,4.]])
plot_gersh(A)
```

2. When A is diagonal, then the circles collapse to points.

```python
A = np.array([[1.,0.0,0.0],[0.0,3.0,0.0],[0.0,0.0,-5.0]])
plot_gersh(A)
```

3. When A is positive definite, the eigenvalues are positive but the circles may cross the imaginary axis.

```python
A = np.array([[2.,-1.,0.0],[-1.0,2.0,-1.0],[0.0,-1.0,2.0]])
plot_gersh(A)

d,v = np.linalg.eig(A)
print("Eigenvalues of A are", d) #eigenvalues are positive
```

Although we have plotted the discs explicitly, we know that the eigenvalues lie on the real axis (hermitian) and are positive (because eigenvalues are positive) so the discs should really be intervals.

4. When $A$ is  strictly diagonally dominant, the discs do not touch the imaginary axis, so all the eigenvalues cannot be zero, so the matrix is invertible. 


$$ A = \begin{bmatrix} -4 & 2 & 1  \\ 1 & 6 & 2  \\ 1 & -2 & 5 \end{bmatrix}.  $$ 
(Example from wikipedia).

```python
A = np.array([[-4.,2.,1.],[1.,6.,2.],[1.,-2.,5.]])
plot_gersh(A)

d,v = np.linalg.eig(A)
print("Eigenvalues of A are", d) #eigenvalues are nonzero
```

5. Suppose $A$ is orthogonal, then the eigenvalues lie on the unit circle.

```python
A = np.random.randn(3,3)
Q,r = np.linalg.qr(A) #Q is orthogonal
plot_gersh(Q)
d,v = np.linalg.eig(Q)
print("Eigenvalues of Q are", d) #Lie on the unit circle
```

5. When $A$ is skew-symmetric, the eigenvalues are purely imaginary.

```python
A = np.array([[0.,-2.,-1.],[2.,0.,4,],[1.,-4.,0.]])
plot_gersh(A)
d,v = np.linalg.eig(A)
print("Eigenvalues of A are", d) #Lie on the imaginary axis
```

### Some ways to strengthen this theorem

1. We can use the discs with radii computed from the columns of the matrix. 
$$ \lambda \in  \bigcup_{j=1}^n D_j^c \qquad D_j^c = \{ z \in \mathbb{C}| |z - a_{ii}| \leq \sum_{i=1, i \neq j}^n |a_{ij} \}  $$
2. Can take the intersection of the row discs and the column discs. 
3. If the union of k discs is disjoint from the union of the other $n − k$ discs then the former union contains exactly $k$ and the latter $n − k$ eigenvalues of A. (Source: Wikipedia).
4. Eigenvalues are invariant to similarity, so use a diagonal similarity to obtain tighter discs.
