### Newton's method for square root


To compute $(\alpha)^{1/2}$ for $\alpha > 0$, we use Newton's method to the system of equations

$$ f(x) = x^2 - \alpha $$

which gives (for $x_0 > 0$)
$$ x_{k+1} = \frac{1}{2}(x_k + \frac{\alpha}{x_k}). $$

```python
alpha = 2
xkm1 = 10

f = lambda x: x**2. - alpha
fx0 = f(xkm1) 
atol = 1.e-10
rtol = 1.e-10



err = []
print('True solution %0.12g' %(np.sqrt(2)))
while True:
    xk = 0.5*(xkm1+alpha/xkm1)
    print('Current iterate %.12g' % (xk))
    
    #Compute the relative error
    err.append(abs(np.sqrt(alpha)-xk)/np.sqrt(alpha))
    
    #Check for convergence
    if abs(f(xk)) <= atol+rtol*abs(fx0):
        break
    xkm1 = xk
```

```python
plt.semilogy(range(len(err)),err, 'k-')
plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Relative error', fontsize = 16)
```

### Newton's method for division


To compute $1/\alpha$ for $\alpha > 0$, we use Newton's method to the system of equations

$$ f(x) = 1/x - \alpha $$

which gives (for $x_0 > 0$)
$$ x_{k+1} = 2x_k- \alpha x_k^2. $$


Here the initial guess matters a little bit. A bad initial guess does not leave to convergence.

```python
alpha = 9.
xkm1 = 2


f = lambda x: 1/x - alpha
fx0 = f(xkm1) 
atol = 1.e-10
rtol = 1.e-10

err = []
print('True solution %0.12g' %(1/alpha))
while True:
    xk = xkm1*(2 - alpha*xkm1)
    print('Current iterate %.12g' % (xk))
    
    #Compute the relative error
    err.append(abs(1/alpha-xk)/abs(1/alpha))
    
    #Check for convergence
    if abs(f(xk)) <= atol+rtol*abs(fx0):
        break
    xkm1 = xk
```

```python
plt.semilogy(range(len(err)),err, 'k-')
plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Relative error', fontsize = 16)
```

### Newton's method for Lambert-w function

We want to solve the Lambert-W equation 

$$ we^w = x \qquad x > 0.$$ 

Newton's method for an initial guess $w_n$

$$ w_{k+1} = w_k \frac{w_k + x/(w_ke^{w_k})}{w_k + 1}. $$

```python
wt = 5

x = wt*np.exp(wt)
wkm1 = 10


f = lambda w: w*np.exp(w) - x
fx0 = f(wkm1) 
atol = 1.e-10
rtol = 1.e-10



err = []
print('True solution %0.12g' %(5))
while True:
    wk = wkm1*(wkm1+x/(wkm1*np.exp(wkm1)))/(wkm1+1)
    print('Current iterate %.12g' % (wk))
    
    #Compute the relative error
    err.append(abs(wt-wk)/abs(x))
    
    #Check for convergence
    if abs(f(wk)) <= atol + rtol*abs(fx0):
        break
    wkm1 = wk
```

```python
plt.semilogy(range(len(err)),err, 'k-')
plt.xlabel('Number of iterations', fontsize = 16)
plt.ylabel('Relative error', fontsize = 16)
```
